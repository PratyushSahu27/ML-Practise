```python
heart_disease = pd.read_csv('13.1 heart-disease.csv')
```

```python
heart_disease[:10]
```

```python
heart_disease['age'].plot.hist();
```

```python
heart_disease['age'].plot.hist(bins=500);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',y='chol',c='target',ax=ax);
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',y='chol',c='target',ax=ax);
ax.set_xlim([45,100])
```

```python
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=heart_disease['age'],
                    y = heart_disease['chol'],
                     c = heart_disease['target'])
ax.set(title='Heart Disease',xlabel='Age',ylabel='Cholestrol')
ax.legend(*scatter.legend_elements(),title='target')
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter = ax1.scatter(x=heart_disease['age'],
                    y = heart_disease['chol'],
                     c = heart_disease['target'])
ax1.set(title='Heart Disease and Cholestrol Levels',xlabel='Age',ylabel='Cholestrol')
ax1.legend(*scatter.legend_elements(),title='target')
ax1.axhline(y = over_50['chol'].mean(), linestyle='--');
scatter = ax2.scatter(x=heart_disease['age'], y= heart_disease['thalach'], c= heart_disease['target'])
ax2.set(title='Heart Disease and Max Heart Rate',xlabel='Age', ylabel='Max Heart rate')
ax2.legend(*scatter.legend_elements(),title='Target')
ax2.axhline(y=over_50['thalach'].mean(),linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=22, fontweight='bold')
```
